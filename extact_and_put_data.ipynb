```python
import os
import psycopg2
from dotenv import load_dotenv
import pandas as pd
```

```python
load_dotenv()
def get_postgresql_connection():
    return psycopg2.connect(
        host=os.getenv('DB_HOST'),
        port=int(os.getenv('DB_PORT', 5432)),
        database=os.getenv('DB_NAME'),
        user=os.getenv('DB_USER'),
        password=os.getenv('DB_PASSWORD')
    )
```

```python
connection = get_postgresql_connection()
```

```python
#cursor = connection.cursor()
retrieve_exam_query = "SELECT  app_examen.id as id, app_examen.nom as nom, app_examen.laboratoire_id as labo_id, app_examen.departement_id as depar_id \
                    FROM app_examen INNER JOIN app_laboratoire ON app_examen.laboratoire_id=app_laboratoire.id \
                    where app_examen.laboratoire_id=2"
retrieve_departement_query = "SELECT app_departement.id as id, app_departement.nom as nom\
                    FROM app_departement"
retrieve_provenance_query = "SELECT app_provenance.id as id, app_provenance.nom as nom\
                    FROM app_provenance\
                    WHERE laboratoire_id=2"
retrieve_indice_query = "SELECT app_indice.id as id, app_indice.nom_complet as nom_complet,\
                    app_indice.nom_affiche as nom_affiche, app_examen.laboratoire_id, app_examen.id as examen_id\
                    FROM app_indice\
                    INNER JOIN app_examen_indices ON app_indice.id = app_examen_indices.indice_id\
                    INNER JOIN app_examen\
                    ON app_examen_indices.examen_id = app_examen.id\
                    WHERE app_examen.laboratoire_id =2"
#retrieve_query = "SELECT * FROM app_examen"
#cursor.execute(retrieve_query)
```
